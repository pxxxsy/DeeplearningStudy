```python
class parent_class:
    prt_v1 = 'a'
    prt_v2 = 0
    prt_v3 = '0'
    share_v = 1

    def show_val(self):
        print('parent:')
        print(self.share_v)

class child_class(parent_class):
    
    def __init__(self, v):
        self.share_v = v

    def show_val(self):
        super().show_val()
        print(super().prt_v1)
        print(super().prt_v2)
        print(super().prt_v3)
        print('child:')
        print(self.share_v)
        
c1 = child_class(4)
c1.show_val()
```

#**Iterator**

Iterable object- list,tuple,set,dictionary,문자열

interable object에서 itertaion을 수행하는 iterator이며 iterator은 next로 다음 원소를 꺼내옴

더이상 반환할 next가 부재할 시 stopiteration 예외상황 반환

```python
list_iterobj = [1,3,5,7,9]
list_iterator = iter(list_iterobj)
print(list_iterator.__next__())#in python2 list_iterator.next()
print(list_iterator.__next__())
print(list_iterator.__next__())
print(next(list_iterator))
print(next(list_iterator))
#print(list_iterator.__next__())#stopiteration exception err occur
```

iterable한 class를 원할 시 iterator를 반환하는 _iter_() 작성

동일한 클래스 객체 혹은 별도로 작성된 iterator 클래스의 객체

iterator가 되는 클래스는 반드시 _next_()를 가짐

요약) iterable class(object)->has iterator->has _next_()

```python
class iterator_class:
    def __init__(self,capacity_,list_):
        self.cap = capacity_
        self.index = 0
        if not list_:
            self.data = list(range(self.cap))
            print(self.data)
        else:
            self.data = list_
            print(self.data)
    def __iter__(self):
        return self
    def __next__(self):
        if self.index >= self.cap:
            raise StopIteration
        out = self.data[self.index]
        self.index += 1
        return out
test = iterator_class(10,[])
for x in test:
    print(x)
######################################################
class iterable_class:
    def __init__(self,capacity_,list_):
        self.cap = capacity_
        self.list = list_
    def __iter__(self):
        self.index = 0
        return iterator_class(self.cap,self.list)
test = iterable_class(5,[1,3,5,7,9])
for x in test:
    print(x)
```

#**Generator**

iterator의 특수한 형태 중 하나

yeild가 사용되어 데이터를 하나씩 리턴

next를 통해 호출될 때마다 마지막 yield 이후부터 다음 yeild까지의 동작 수행

for문에서 사용

```python
def generator_func():
    print('0-1')
    yield 1
    print('1-2')
    yield 2
    print('2-3')
    yield 3
    print('3-4')

g1 = generator_func()
print(type(g1))
print(next(g1))
print(next(g1))
print(next(g1))
print('--------------------')
g2 = generator_func()
for item in g2:
    print(item)
```

```python
def infgen():
    cnt = 0
    while True:
        yield cnt
        cnt = cnt + 1
        
g = infgen()
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

예외상황을 직접 만들고 싶은 경우 class를 만들고 raise를 통해 예외상황 발생시 예외상황임을 알림

```python
class myerr(Exception):
    pass

def errcallfunction(arg):
    if arg == True:
        raise myerr()
    print(arg)
    
try:
    errcallfunction(False)
    errcallfunction(True)
except myerr:
    print('err ouccured!')
```

```python
class myerr(Exception):
    def __str__(self):
        return "err message!"

def errcallfunction(arg):
    if arg == True:
        raise myerr()
    print(arg)
    
try:
    errcallfunction(False)
    errcallfunction(True)
except myerr as E:
    print('err ouccured!')
    print(E)
```
